```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,5,100)
y1= 3+np.exp(1)-x*np.sin(6*x)
y2= 4+np.exp(-x)*np.cos(6*x)

fig,ax=plt.subplots()
ax.plot(x,y1,color="blue", label="y(x)")
ax.plot(x,y2,color="red", label="y`(x)")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.savefig('201810569 김정범.pdf')
```
